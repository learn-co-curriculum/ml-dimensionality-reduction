# 2. DATA PREP

```python
import pandas as pd
import numpy as np
absence = pd.read_csv("Absenteeism_at_work_2.csv", sep = ';')
import matplotlib.pyplot as plt
absence.head()
```

```python
absence["Work load per day"] = absence["Work load per day"].replace(
    [246.288, 241.476, 251.818, 236.629, 377.550, 313.532, 265.017, 294.217, 284.853, 246.074],
    [np.nan, np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan ])
absence["Service time"] = absence["Service time"].replace(
    [3,6,7,8,9,11,13, 18, 22, 29],
    [np.nan, np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan ])
```

```python
df1 = absence.loc[:, ['Reason for absence', 'Month of absence', 'Day of the week', 
                      'Seasons','Transportation expense', 'Distance to Work', 'Service time', 'Age', 'Work load per day', 
                     'Education','Son', 'Social drinker', 'Social smoker', 'Weight', 'Height', 'Body mass index', 
                     'Absenteeism time in hours']]
```

```python
df1.to_csv("absence_data.csv")
```
